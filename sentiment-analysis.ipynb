# Twitter sentiment analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('tweets-wilders.csv', parse_dates=['date'])
df.is_retweet=df.is_retweet.apply(lambda x: (x and 'r') or 'b')
plt.scatter(df.polarity, df.subjectivity, alpha=0.1, c=df.is_retweet)
```
